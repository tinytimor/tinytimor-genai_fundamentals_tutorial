```python
def drop_empty_dfs(dfs):
    """
    Drop empty DataFrames from a list of DataFrames.

    Args:
        dfs (list): A list of pandas DataFrames.

    Returns:
        list: A list of non-empty pandas DataFrames.
    """
    non_empty_dfs = []
    for df in dfs:
        if not df.empty:
            non_empty_dfs.append(df)
    return non_empty_dfs
```

```python
pattern = r"\[START_\d+\]\s*(.*?)\s*\[END_\d+\]"
queries = re.findall(pattern, string_response, re.DOTALL)
for i in queries:
    print(i)
df_ls = [pd.read_sql_query(query, conn) for query in queries]
dfs = drop_empty_dfs(df_ls)
```
